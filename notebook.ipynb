Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
best_math_schools = pd.DataFrame(schools[["school_name","average_math"]])
best_math_schools = best_math_schools[best_math_schools['average_math']>=(.8*800)]
best_math_schools.sort_values("average_math", ascending=False, inplace = True )
print(best_math_schools)
```

Here are the best maths schools that are arranged in the basis of a descending highest to lowest average math scores

```python
top_10_schools =pd.DataFrame(schools[["school_name"]])
#note: using axis =1 means rowwise sum while by default the axis will be 0
top_10_schools["total_SAT"] = schools[["average_math","average_reading","average_writing"]].sum(axis=1)
top_10_schools = top_10_schools.sort_values("total_SAT",ascending= False )
top_10_schools = top_10_schools.iloc[:10]
#top_10_schools.set_index("school_name",inplace=True)

print(top_10_schools)
```

Here are the top 10 most highest SAT scoring Schools 

```python
schools["total_SAT"] = schools[["average_math","average_reading","average_writing"]].sum(axis=1)
schools.head()

import numpy as np
```

```python



grouped_borough = schools.groupby("borough").agg( {
    'school_name': 'size',
    'total_SAT': ['mean','std']
}).round(2)
grouped_borough.sort_values(by=("total_SAT","std"),ascending=False,inplace=True)

grouped_borough.columns = ['_'.join(col) for col in grouped_borough.columns]
grouped_borough.reset_index(inplace=True)
grouped_borough.rename(columns = {
    'school_name_size':'num_schools',
    'total_SAT_mean':'average_SAT',
    'total_SAT_std':'std_SAT'
},inplace=True)

largest_std_dev = pd.DataFrame(grouped_borough.iloc[[0]])


print(largest_std_dev)
```
